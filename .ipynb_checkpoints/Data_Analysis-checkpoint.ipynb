# DATA ANALYSIS
The main aim of data analysis is to get a better understanding of the dataset we are dealing with. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
## Display all the columns of the dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv("train.csv")
print(dataset.shape) ##(No. of data, No. of features)
dataset.head()
```

### MISSING VALUES

```python
feature_missing = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]
print(feature_missing)
for feature in feature_missing:
    a = np.round(dataset[feature].isnull().mean(), 3)
    print(feature, 'Percentage of missing value:{}%'.format(a))
```

### Plotting the missing values with dependent y variable to find out the relations
We make bar plots where if the data is missing we assign 1, else 0 to get an idea of how much missing values affect certain features 

```python
for feature in feature_missing:
    data = dataset.copy()
    data[feature] = np.where(data[feature].isnull(), 1, 0) #Basically means if it has value, then assign 0 or else assign 1
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

### LOGARITHMIC TRANSFORMATION
In order to implement machine learning algorithms, we need to bring down the continuous variable values. We do it by applying logarithmic scale. 

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()    
```

### 2. Datetime Variables
Here there are four. And each of those features has "Year" or "Yr" in the column heading. We will use that to seperate such temporal data. After that, we analyze the year related features wrt SalePrice to see its effects

```python
temporal_data = [feature for feature in dataset.columns if "Year" in feature or "Yr" in feature]
print(temporal_data)
```

```python
# ANALYZE THE YEAR FEATURES WRT TO SALE PRICE
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")

# dataset.groupby('YrSold')['SalePrice'].median().plot.bar()
# plt.xlabel('Year Sold')
# plt.ylabel('Median House Price')
# plt.title("House Price vs YearSold")
```

```python
#COMPARED THE DIFFERENCE BETWEEN REMAINING THREE FEATURES AND YRSOLD FEATURE WITH OUR Y VARIABLE "SALEPRICE"
for feature in temporal_data:
    if feature!='YrSold':
        data=dataset.copy()
        data[feature]=data['YrSold']-data[feature]

        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```
